```python
import pandas as pd
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2':range(3)})
df1
```

```python
df2
```

# JOIN (by default, INNER)

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

# JOIN with Right Key and Left Key

```python
df3 = pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df4 = pd.DataFrame({'rkey':['a','b','d'],
                   'data2':range(3)})
df3
```

```python
df4
```

```python
pd.merge(df3,df4, left_on = 'lkey', right_on = 'rkey')
```

# Outer Join

```python
pd.merge(df1, df2, how = 'outer')
```

# Left Join

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],
                   'data1':range(6)})
df2 = pd.DataFrame({'key':['a','b','a','b','d'],
                   'data2':range(5)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how = 'left')
```

# Join on 2 keys

```python
left = pd.DataFrame({'key1':['foo','foo','bar'],
                    'key2':['one','two','one'],
                    'lval':[1,2,3]})
right = pd.DataFrame({'key1':['foo','foo','bar','bar'],
                     'key2':['one','one','one','two'],
                     'rval':[4,5,6,7]})
pd.merge(left, right, on=['key1','key2'], how = 'outer')
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=['key1','key2'], how='inner')
```

# Concatenation = Union and Union all in SQL
PUT ONE BELOW THE OTHERS

```python
s1 = pd.Series([0,1], index = ['a','b'])
s1
```

```python
s2 = pd.Series([2,3,4], index = ['c','d','e'])
s2
```

```python
s3 = pd.Series([5,6], index = ['f','g'])
s3
```

```python
pd.concat([s1,s2,s3])
```

Concat with axis = 1 -> puts one below the others, but also creates columns

```python
pd.concat([s1,s2,s3], axis=1)
```

# STACK
If we have 2 indexes, we can take 1 and put it on the top

```python
import numpy as np
data = pd.DataFrame(np.arange(6).reshape(2,3), 
                   index = pd.Index(['Ohio','Colorado'], name='state'),
                   columns = pd.Index(['one','two','three'], name = 'number'))
data
```

```python
result = data.stack()     # one, two, three come down
result
```

```python
result.unstack()
```

```python
df = pd.DataFrame({'left':result,'right': result + 5},
                 columns = pd.Index(['left','right'], name = 'side'))
df
```

```python
df.unstack()
```
